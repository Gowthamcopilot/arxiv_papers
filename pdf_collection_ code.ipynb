```python
import random

def generate_custom_headers():
    accept_options = [
    "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "text/html,application/json;q=0.9,*/*;q=0.8",
    "text/html,application/xhtml+xml;q=0.9,*/*;q=0.8"
    ]


    accept_language_options = [
        "en-US,en;q=0.9",
        "en-GB,en;q=0.8",
        "en;q=0.7",
        "en-US,en;q=0.9,fr;q=0.8",
    ]

    accept_encoding_options = [
        "gzip, deflate, br",
        "gzip, deflate",
    ]

    referer_options = [
    "https://www.google.com/",
    "https://www.bing.com/",
    "https://news.google.com/",
    "https://www.yahoo.com/",
    "https://duckduckgo.com/",
    "https://www.facebook.com/",
    "https://twitter.com/",
    "https://www.reddit.com/",
    "https://www.youtube.com/",
    ]


    upgrade_insecure_requests_options = [
        "1",
    ]

    sec_fetch_mode_options = [
        "navigate"
    ]

    headers = {
        "User-Agent" : agent.random,
        "Accept": random.choice(accept_options),
        "Accept-Language": random.choice(accept_language_options),
        "Accept-Encoding": random.choice(accept_encoding_options),
        "Connection": 'keep-alive',
        "Referer": random.choice(referer_options),
        "Upgrade-Insecure-Requests": "1",
        "Sec-Fetch-Dest": "document",
        "Sec-Fetch-Mode": random.choice(sec_fetch_mode_options),
        "Sec-Fetch-Site": "cross-site",
        "Sec-Fetch-User": "?1",
    }

    return headers
agent = UserAgent()
```

```python

import pandas as pd
df = pd.read_csv('/content/arxiv_cs_metadata.csv') # replace with your own path of csv file of papers
df['is_scraped'] = False
df['pdf_link'] = ''
for i in range(0, len(df)):
    df.loc[i, 'pdf_link'] = 'https://arxiv.org/pdf/' + str(df.loc[i, 'id']) + '.pdf'
```

```python
df
```

```python
import requests
import time
from tqdm import tqdm
import os

# Directory where PDFs will be saved
save_directory = '/content/files/'
os.makedirs(save_directory, exist_ok=True)

def download_pdf(url, save_path):
    headers = generate_custom_headers()
    try:
        response = requests.get(url, headers=headers, stream=True)
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=1024):
                    if chunk:
                        f.write(chunk)
            print(f"Downloaded: {save_path}")
            return True
        else:
            print(f"Failed to download {url} - Status Code: {response.status_code}")
            return False
    except Exception as e:
        print(f"Error downloading {url}: {e}")
        return False

# Filter only unscraped entries
unscraped_df = df[df['is_scraped'] == False].copy()

batch_size = 500
total_unscraped = unscraped_df.shape[0]
num_batches = (total_unscraped + batch_size - 1) // batch_size

for batch_num in range(num_batches):
    start_idx = batch_num * batch_size
    end_idx = min(start_idx + batch_size, total_unscraped)
    batch_df = unscraped_df.iloc[start_idx:end_idx]

    print(f"\nProcessing batch {batch_num + 1}/{num_batches} ({start_idx} to {end_idx})")

    for index, row in tqdm(batch_df.iterrows(), total=batch_df.shape[0], desc=f"Batch {batch_num + 1}"):
        url = row['pdf_link']
        pdf_id = row['title'].replace('/', '_').replace('\\', '_')
        save_path = f"{save_directory}{pdf_id}.pdf"

        # Download PDF
        success = download_pdf(url, save_path)

        # Update is_scraped if successful
        if success:
            df.at[row.name, 'is_scraped'] = True

        time.sleep(3)  # Delay

    # Ask whether to continue
    user_input = input(f"\nBatch {batch_num + 1} completed. Continue to next batch? (yes/no): ").strip().lower()
    if user_input != 'yes':
        break

# Save updated DataFrame
df.to_csv("updated_df.csv", index=False)
print("Scraping session ended. Progress saved to 'updated_df.csv'.")
```

```python
import zipfile
import os

# Path to your folder containing the PDFs
folder_path = '/content/files/'

# Zip the files in that folder
zip_filename = '/content/files.zip'
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            zipf.write(os.path.join(root, file), arcname=file)

print(f"All files zipped as {zip_filename}")
```
